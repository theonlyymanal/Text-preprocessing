```python
import re
import string
import nltk
import pandas as pd

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    #Je convertis au minuscule:
    text = text.lower()
    #J'enlève les ponctuations:
    text = text.translate(str.maketrans('', '', string.punctuation))
    #Je decoupe les mots:
    words = text.split()
    #Je supprime les mots vides :
    custom_stopwords = set(["is", "i", "the", "in", "so", "be", "a", "for"])
    text = ' '.join([word for word in text.split() if word.lower() not in custom_stopwords])
    #Stemming
    stemmer = PorterStemmer()
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    #Lemmatization
    lemmatizer = WordNetLemmatizer()
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
    #Et en fin je supprime les émojis et les URL:
    text = re.sub(r'<.*?>|:\)|:\(|:\D|<3', '', text)
    text = re.sub(r'http\S+', '', text)

    return text
```
